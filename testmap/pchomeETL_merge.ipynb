```python
from bs4 import BeautifulSoup
import requests as res
from datetime import datetime
import csv
import time 
import multiprocessing as mp
import random
```

```python
def pchome_product(url):
    pchome_product_res=res.get(url)
    time.sleep(1)
    pchome_product_soup = BeautifulSoup(pchome_product_res.text,'html.parser') #使用html.parser格式做soup
    businessmen=pchome_product_soup.select(' a > div')[0].text.replace(",","").replace('"',"")
    name=pchome_product_soup.select(' div.info > div.tit')[0].text.replace(",","").replace('"',"")
    Freight=pchome_product_soup.select('div.info > ul > li > span')[1].text.replace(",","").replace('"',"")
    price=pchome_product_soup.select(' div.info > ul > li > span')[0].text.replace(",","").replace('"',"").replace(' ',"")
    if str(price) != "元":    
        #寫出csv檔
        f = open("pchome.csv","a")
        f.writelines(businessmen + "," + str(datetime.now()) + "," +name+ "," + price+ "," + url+ "," +Freight+ "," +"\n")
        f.close()
```

```python
def pchome_product2(url):
    pchome_product_res=res.get(url)
    time.sleep(1)
    pchome_product_res.encoding = 'big5'
    pchome_product_soup = BeautifulSoup(pchome_product_res.text,'html.parser') #使用html.parser格式做soup
    businessmen=pchome_product_soup.select('  tr > td > a')[10].text.replace(",","").replace('"',"")
    name=pchome_product_soup.select(' td > h1 ')[0].text.replace(",","").replace('"',"")
    price=pchome_product_soup.select('  tr > td > span')[3].text.replace(",","").replace('"',"").replace(' ',"")
    if str(price) != "元":
        f = open("pchome.csv","a")
        f.writelines(businessmen + "," + str(datetime.now()) + "," +name+ "," + price+ "," + url+ "," + "," +"\n")
        f.close()
    
```

```python
def getProductLinks(url,pool):
    pchome_res=res.get(url)
    pchome_res.encoding = 'big5'  #改成網頁big5的編碼
    pchome_soup = BeautifulSoup(pchome_res.text,'html.parser') #使用html.parser格式做soup
    #取得個別商品連結
    n=0
    
    product_links=pchome_soup.select('div.pic2.pic2-sclass > div.pic2t.pic2t_bg > a')
    for product_link in product_links: 
        if product_link['href'][0] == "/":
            product_link['href']="http://www.pcstore.com.tw"+product_link['href']
            #pchome_product2(str(product_link['href']))
            pool.apply_async(pchome_product2, (str(product_link['href']), ))
        else:
            pool.apply_async(pchome_product, (str(product_link['href']), ))
            #pchome_product(str(product_link['href']))
        print(".",end="")
        time.sleep(random.random()/3)
        n+=1
    print('\n')
    print(n)
```
